```python
import face_recognition
import cv2
from google.colab.patches import cv2_imshow
```

```python
x = face_recognition.load_image_file('/content/v1.jpg')
virat_encoding= face_recognition.face_encodings(x)[0]
known_encoding=[virat_encoding]
known_name=['virat']
```

```python
known_encoding
```

```python
y = face_recognition.load_image_file('/content/v2.jpg')
x_enc1=face_recognition.face_encodings(y)[0]
print(x_enc1)
```

```python
results=face_recognition.compare_faces(known_encoding, x_enc1)
results
```

```python
for i in range(len(results)):
  if results[i]:
    name= known_name[i]
    (top,right,bottom,left)=face_recognition.face_locations(y)[0]
    cv2.rectangle(y,(left,top),(right,bottom),(0, 0, 255), 2)
    cv2.putText(y,name,(left+2,bottom+20),cv2.FONT_HERSHEY_PLAIN,1,(225, 225, 225), 1)
    cv2_imshow(y)
```
